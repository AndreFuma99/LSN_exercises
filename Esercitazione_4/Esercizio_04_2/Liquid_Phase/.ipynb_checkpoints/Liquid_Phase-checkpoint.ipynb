```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

```python
plt.rcParams.update({'font.size': 18})
```

```python
data1 = np.loadtxt("/home/andrea/Documenti/CorsoLSN/Esercitazione_4/Esercizio_04_2/Liquid_Phase/nsl_simulator/NSL_SIMULATOR/OUTPUT/total_energy.dat", skiprows=1)
```

```python
N=10                # Number of blocks
L=1000            # Number of steps in each block
x = np.arange(N)      # [0,1,2,...,N-1]

x*=L # Number of steps = block * (Number of steps in each block)
plt.errorbar(x,data1[:, 2],yerr=data1[:, -1])
plt.xlabel('#steps')
plt.ylabel('E*/N')
plt.title('Total (internal) energy per particle')
plt.grid(True)
plt.show()
```
