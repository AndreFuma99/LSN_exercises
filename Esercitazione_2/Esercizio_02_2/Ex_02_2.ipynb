```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

```python
data1 = pd.read_csv("Es02_2a.csv", sep=',')
```

```python
print(data1.head(8))
```

```python
M=10000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N)      # [0,1,2,...,N-1]

plt.errorbar(x,data1['CumulativeAverage'],yerr=data1['StatisticalUncertainty'])
plt.xlabel('#walks')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.grid(True)
plt.show()
```

```python
data2 = pd.read_csv("Es02_2b.csv", sep=',')
```

```python
print(data2.head(8))
```

```python
plt.errorbar(x,data2['CumulativeAverage'],yerr=data2['StatisticalUncertainty'])
plt.ylim(0, 10)
plt.xlabel('#walks')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.grid(True)
plt.show()
```

```python
plt.rcParams.update({'font.size': 18})
plt.title('3D RW on a cubic lattice')
plt.xlabel('#walks')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.errorbar(x,data1['CumulativeAverage'],yerr=data1['StatisticalUncertainty'])
plt.plot(x, 0.94*np.sqrt(x), color="red")
plt.grid(True)
plt.show()
```
