### Esercizio 01.1

In questo esercizio voglio testare il Parallel Random Number Generator, ossia il generatore di numeri pseudo-casuali che abbiamo utilizzato durante il corso.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
from scipy.stats import chi2, norm, cauchy
```

Per prima cosa voglio calcolare il valor medio dei numeri generati dalla routine RANdom New York University (**RANNYU**) e confrontarlo con la media della distribuzione uniforme.
Infatti, per il momento, posso considerare come ipotesi che questi numeri siano tratti da una distribuzione uniforme.

```python
data1 = pd.read_csv("Esercizio_01_1/Es01_1_01.csv", sep=',')
```

```python
M=100000              # Numero totale di lanci
N=100                 # Numero di blocchi
L=int(M/N)            # Numero di lanci in ogni blocco
x = np.arange(N)      # [0,1,2,...,N-1]

x*=L # Numero di lanci = blocco * (Numero di lanci in ogni blocco)
plt.errorbar(x,data1['Cumulative_Average']-1/2,yerr=data1['Statistical_Uncertainty'])
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

In questo plot ho mostrato l'andamento della differenza tra il valor medio di r ($\langle r \rangle$) e la media della distribuzione uniforme tra 0 e 1 (1/2), in funzione del numero totale di "lanci". Questo andamento, che dovrebbe stabilizzarsi all'aumentare del numero di lanci a 0, è stato ottenuto tramite il **metodo della media a blocchi**: quello che vado a plottare è quindi in realtà la media cumulativa sui blocchi.
Il metodo della media a blocchi mi ha consentito di associare ad ogni stima di $\langle r \rangle$ un'incertezza statistica (deviazione standard della media), rappresentata dalla stima della deviazione standard $\sigma$ (radice quadrata del momento centrale secondo) divisa per $\sqrt{N-1}$, con $N$ numero di blocchi considerato e $r_i$ numeri pseudo-casuali generati::

$$\frac{\sigma}{\sqrt{N-1}} \simeq \sqrt{\frac{1}{N-1} \left[ \frac{1}{N} \sum_{i=1}^N r^2_i - \left( \frac{1}{N} \sum_{i=1}^N r_i \right)^2 \right]} $$

Un ulteriore test che si può effettuare consiste nel calcolo del momento centrale secondo dei numeri pseudo-casuali e nel confronto di questo con la varianza della distribuzione uniforme.

```python
data2 = pd.read_csv("Esercizio_01_1/Es01_1_02.csv", sep=',')
```

```python
plt.errorbar(x,data2['Cumulative_Average']-1/12,yerr=data2['Statistical_Uncertainty'])
plt.xlabel('#throws')
plt.ylabel('$<(r-1/2)^2> - 1/12$')
plt.grid(True)
plt.show()
```

In questo plot ho mostrato l'andamento della differenza tra il momento centrale secondo di r ($\langle (r-1/2)^2 \rangle$) e la varianza della distribuzione uniforme tra 0 e 1 (1/12), in funzione del numero totale di "lanci". Come prima, questo andamento è stato ottenuto tramite il metodo della media a blocchi; l'incertezza statistica corrisponde alla deviazione standard della media per la stima di $\langle (r-1/2)^2 \rangle$.

Come ho detto prima, l'ipotesi che sta alla base è che questi numeri pseudo-casuali r siano tratti da una distribuzione uniforme. In realtà, a livello statistico, non ha senso dire che degli eventi casuali siano tratti da una particolare distribuzione. Provo quindi a testare quest'ipotesi.

```python
data3 = pd.read_csv("Esercizio_01_1/Es01_1_03.csv", sep=',')
```

```python
plt.hist(data3['Chi_Square'], bins=20, density=True, alpha=0.75, label='Observed Chi-square')
x = np.linspace(min(data3['Chi_Square']), max(data3['Chi_Square']), 1000)
plt.plot(x, chi2.pdf(x, df=N-1), 'r-', label=f'Chi-Square Distribution')
plt.xlabel('$\chi^2$')
plt.ylabel('Frequency')
plt.legend().get_frame().set_facecolor((1, 1, 1, 0.5))
plt.title('Chi-Square Test for Uniform Distribution')
plt.show()
```

In questo plot ho riportato tramite un istogramma, in blu, i valori di $\chi^2$ "osservati"; in rosso invece ho rappresentato l'andamento della vera distribuzione di probabilità del $\chi^2$. Se i numeri generati pseudo-casualmente fossero tratti da una distribuzione uniforme, mi aspetterei $\chi^2 \simeq 100$ (quasi il picco dell'andamento rosso), ossia circa pari al rapporto tra il numero di "lanci" ad ogni tentativo e il numero di sotto-intervalli in cui ho diviso [0,1]. In quanto questo istogramma mostra due picchi ben maggiori di 100, posso rigettare l'ipotesi che i numeri siano tratti da una distribuzione uniforme: questo è sensato, perchè i numeri psedo-casuali sono in realtà tratti da un algoritmo deterministico.

### Esercizio 01.2

L'obiettivo è quello di estendere il generatore di numeri pseudo-casuali in modo da verificare il **Teorema del Limite Centrale**, il quale afferma che, data una popolazione con una qualsiasi distribuzione di probabilità, <font color="red">i cui momenti primo e secondo non divergono</font>, la distribuzione della somma (o della media) di un gran numero di campioni indipendenti e identicamente distribuiti estratti da questa popolazione tende a seguire una distribuzione normale, indipendentemente dalla distribuzione originale della popolazione.

Per fare questo ho utilizzato il **metodo dell'inversa della distribuzione cumulativa** per campionare una generica distribuzione esponenziale, $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$, e una generica distribuzione di Cauchy-Lorentz, $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$.

```python
data4 = pd.read_csv("Esercizio_01_2/Es01_2_std.csv", sep=',')
```

```python
plt.hist(data4['Standard_Sums'], bins=30, density=True, alpha=0.75, label='Standard dice')
x = np.linspace(min(data4['Standard_Sums']), max(data4['Standard_Sums']), 1000)
mu, std = norm.fit(data4['Standard_Sums'])
plt.plot(x, norm.pdf(x, mu, std), 'r-', label=f'Gaussian Distribution')
plt.xlabel('$S_{100}$')
plt.ylabel('Frequency')
plt.legend()
plt.title('Fit with a Gaussian for Standard dice')
plt.show()
```

In questo plot ho rappresentato il fit con una distribuzione Gaussiana per un dado Standard. Per costruire l'istogramma, in blu, ho calcolato 10000 volte $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, con $N=100$, essendo $x_i$ delle variabili casuali campionate lanciando un dado Standard. In rosso invece ho presentato il fit con una distribuzione Gaussiana. E' possibile osservare che il Teorema del Limite Centrale viene rispettato, in quanto la distribuzione della media tende a seguire la distribuzione Gaussiana.

```python
data5 = pd.read_csv("Esercizio_01_2/Es01_2_exp.csv", sep=',')
```

```python
plt.hist(data5['Exponential_Sums'], bins=30, density=True, alpha=0.75, label='Exponential dice')
x = np.linspace(min(data5['Exponential_Sums']), max(data5['Exponential_Sums']), 1000)
mu, std = norm.fit(data5['Exponential_Sums'])
plt.plot(x, norm.pdf(x, mu, std), 'r-', label=f'Gaussian Distribution')
plt.xlabel('$S_{100}$')
plt.ylabel('Frequency')
plt.legend()
plt.title('Fit with a Gaussian for Exponential dice')
plt.show()
```

In questo plot ho rappresentato il fit con una distribuzione Gaussiana per un dado Esponenziale. Per costruire l'istogramma, in blu, ho calcolato 10000 volte $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, con $N=100$, essendo $x_i$ delle variabili casuali campionate lanciando un dado Esponenziale. In rosso invece ho presentato il fit con una distribuzione Gaussiana. E' possibile osservare che il Teorema del Limite Centrale viene rispettato, in quanto la distribuzione della media tende a seguire la distribuzione Gaussiana.

Questi due fit con una distribuzione Gasussiana mostrano anche che il Teorema del Limite Centrale viene rispettato indipendentemente dalla distribuzione di partenza.

Nel caso di un dado Lorentziano, la variabile casuale segue una distribuzione di Cauchy-Lorentz. Poichè però <font color="red">la varianza di una distribuzione di Cauchy-Lorentz non è finita</font>, non è possibile applicare il Teorema del Limite Centrale: la distribuzione della media non converge ad una distribuzione Gaussiana.

```python
data6 = pd.read_csv("Esercizio_01_2/Es01_2_lor.csv", sep=',')
```

```python
mask = (data6['Lorentzian_Sums'] >= -15) & (data6['Lorentzian_Sums'] <= 15)
```

Il mio obiettivo è quello di fittare con una distribuzione di Cuachy-Lorentz per un dado Lorentziano. Per effettuare un fit decente rimuovo dai 10000 valori di $S_{100}$ tutti quelli con valore assoluto maggiore di 15, in quanto possiedono una frequenza molto prossima allo 0, e quindi trascurabile.

```python
filtered_data6 = data6['Lorentzian_Sums'][mask]
```

```python
plt.hist(filtered_data6, bins=30, density=True, alpha=0.75, label='Lorentzian dice')
x = np.linspace(min(filtered_data6), max(filtered_data6), 1000)
loc, scale = cauchy.fit(filtered_data6)
plt.plot(x, cauchy.pdf(x, loc, scale), 'r-', label=f'Cauchy Distribution')
plt.xlabel('$S_{100}$')
plt.ylabel('Frequency')
plt.legend()
#plt.xlim(-500, 500)
#plt.ylim(0, 0.003)
plt.title('Fit with a Cacuchy-Lorentz for Lorentzian dice')
plt.show()
```

In questo plot ho rappresentato il fit con una distribuzione di Cauchy-Lorentz per un dado Lorentziano. Per costruire l'istogramma, in blu, ho calcolato 10000 volte $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, con $N=100$, essendo $x_i$ delle variabili casuali campionate lanciando un dado Lorentziano. In rosso invece ho presentato il fit con una distribuzione di Cauchy-lorentz. In quanto la distribuzione di Cauchy-Lorentz, possedendo una lunga coda che segue una legge di potenza inversa, è una **distribuzione stabile**, è invariante sotto la somma e la media. Questo significa che la media di variabili casuali estratte lanciando un dado Lorentziano è ancora una variabile casuale di Cauchy-Lorentz, come si evince dal fit.

### Esercizio 01.3

L'obiettivo di questo esercizio è quello di simulare **l'esperimento di Buffon**: un ago di lunghezza $L$ viene lanciato casualmente su un piano orizzontale segnato da linee parallele a distanza $d$ l'una dall'altra. La probabilità $P$ che l'ago intersechi una di queste linee è: $P = 2L/\pi d$ (ho scelto $d=2>L=1$ ma non $d>>L$ poichè implicherebbe $P<<1$). Questo dovrebbe essere usato per stimare $\pi$ dai lanci dell'ago. Se l'ago è lanciato $N_{thr}$ volte ed è osservato fermarsi sopra una linea $N_{hit}$ di queste volte, posso stimare $\pi$ da:
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$

```python
data7 = pd.read_csv("Esercizio_01_3/Es01_3.csv", sep=',')
```

```python
M=100000              # Numero totale di trials = Numero di lanci dell'ago per ogni trial
N=100                 # Numero di blocchi
L=int(M/N)            # Numero di trials in ogni blocco
x = np.arange(N)      # [0,1,2,...,N-1]

x*=L # Numero di trials = blocco * (Numero di trials in ogni blocco)
plt.errorbar(x,data7['Cumulative_Average'],yerr=data7['Statistical_Uncertainty'],label='Observed value of $\pi$')
plt.axhline(y=math.pi, color='red', linestyle='-', linewidth=2, label='Theoretical value of $\pi$')
plt.xlabel('#trials')
plt.ylabel('$<\pi>$')
plt.title('Buffon Experiment')
plt.legend()
plt.grid(True)
plt.show()
```

In questo plot ho mostrato in blu l'andamento del valore stimato di $\pi$ traimite l'esperimento di Buffon, in funzione del numero totale di trials (tentativi), mentre in rosso ho mostrato il valore teorico di $\pi=3.14159$.
L'andamento blu è stato ottenuto tramite il metodo della media a blocchi; l'incertezza statistica corrisponde alla deviazione standard della media per la stima di $\pi$.
