## <span style="color:blue">  Laboratorio di Simulazione Numerica - Esercitazione 08</span>

L'obiettivo di questa esercitazione è quello di ottimizzare variazionalmente lo stato fondamentale di una singola particella quantistica in uno spazio unidimensionale confinata dal seguente potenziale esterno:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

Questo modello quantistico 1D non è risolvibile analiticamente e dovrebbe essere affrontato con metodi numerici. In modo da ottenere una funzione d'onda approssimata per lo stato fondamentale, il **metodo Monte Carlo Variazionale** utilizza una funzione d'onda di trial, parametrizzata da un set di parametri variazionali. Un modello approssimato per la funzione d'onda dello stato fondamentale è la seguente funzione d'onda di trial:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

$\Psi_T^{\sigma,\mu}(x)$ dipende da due parametri variazionali: $\sigma$ e $\mu$.

### Esercizio 08.1

In questo esercizio ho scritto un codice VMC per una singola particella quantistica in 1D che sfrutta l'algoritmo di Metropolis per campionare il modulo quadro della nostra funzione d'onda di trial $|\Psi_T^{\sigma,\mu}(x)|^2$ utilizzando una probabilità di transizione uniforme $T(x_{new}|x_{old})$. Il codice è in grado di calcolare il valore di aspettazione dell'Hamiltoniana

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

che può essere usato per trovare i parametri che minimizzano questa quantità. Ho usato $\hbar=1$ e $m=1$. Per misurare $\langle {\hat H} \rangle_T$ mi sono ricavato l'espressione analitica di $\frac{\partial^2 \Psi_T(x)}{\partial x^2}$:

$$
\frac{\partial^2 \Psi_T(x)}{\partial x^2} =  
\frac{(x-\mu)^2 - \sigma^2}{\sigma^4}e^{-\frac{(x-\mu)^2}{2\sigma^2}} + \frac{(x+\mu)^2 - \sigma^2}{\sigma^4}e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$


### Esercizio 08.2

In questo esercizio, in modo da trovare i parametri $\sigma$ e $\mu$ che minimizzano $\langle {\hat H} \rangle_T$, ho esteso il mio codice VMC per affrontare questo problema di ottimizzazione variazionale utilizzando un **algoritmo di Simulated Annealing**. Al cuore di questo algoritmo c'è un'analogia con la termodinamica, in particolare con il modo in cui i liquidi congelano e cristallizzano. Nel Simulated Annealing uno parte con una simulazione ad alta temperatura e gradualmente raffredda il sistema; se l'ergodicità non è persa durante questo raffreddamento programmato, la simulazione si fermerà nel minimo globale dell'energia: l'essenza del processo è quindi un raffreddamento lento.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

```python
data = pd.read_csv("Esercizio_08_2/SA_optimization.csv")
```

```python
DATA = pd.read_csv("Esercizio_08_2/|psi|2.csv")
```

```python
x = np.linspace(-M, M, 2*M)
m = 0.831713
s = 0.647098

mod_quad = (np.exp(-((x-m)*(x-m))/(s*s)) + np.exp(-((x+m)*(x+m))/(s*s)) + 2*np.exp(-(x*x+m*m)/(s*s))) / (2*s*np.sqrt(np.pi)*(1+np.exp(-(m*m)/(s*s))))



def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations
X = np.linspace(-a/2, a/2, N)
dX = X[1] - X[0] # the step size
V = Vpot(X)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dX**2) + np.diag(V)
E,psi = np.linalg.eigh(H)
psi = np.transpose(psi)
psi = psi/np.sqrt(dX)

print("Ground state energy: ", E[0])



plt.figure(figsize=(12, 6))
plt.hist(DATA, bins=50, density=True, alpha=0.75, label='Sampling')
plt.plot(x, mod_quad, label='Analytic Curve')
plt.plot(X,(psi[0])**2, label='NS by matrix equation')
plt.xlabel('x')
plt.ylabel(r'$|\Psi_T(x)|^2 / \int dx |\Psi_T(x)|^2$')
plt.title('Sampling vs Analytic Curve')
plt.xlim(-2.5, 2.5)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
