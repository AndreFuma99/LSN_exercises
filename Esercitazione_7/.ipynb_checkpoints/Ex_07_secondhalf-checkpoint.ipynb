### Esercizio 07.3

In questo esercizio ho incluso all'interno del codice il calcolo della **funzione di distribuzione radiale** $g(r)$, in funzione della distanza radiale $r$ tra le particelle, con $r \in \left[ 0; L/2\right]$.

La funzione di distribuzione radiale $g(r)$ è una funzione che descrive come varia la densità di particelle in funzione della distanza radiale $r$ da una particella di riferimento. In altre parole, $g(r)$ misura la probabilità di trovare una particella ad una distanza $r$ da un'altra particella rispetto alla probabilità uniforme.
Questa funzione fornisce informazioni dettagliate sulla struttura microscopica del sistema e aiuta a comprendere le forze di interazione tra le particelle.

In pratica, durante la simulazione, ho riempito un'istogramma incrementando il bin di $2$ ad $r$ ogni volta che viene trovata una particella ad una distanza tra $r$ e $r+dr$. Alla fine della simulazione ho normalizzato l'istogramma con la quantità $\rho N \Delta V(r)$, con $\Delta V(r) = \frac{4 \pi}{3} [(r+dr)^3 - r^3]$, che rappresenta il volume dello strato sferico tra $r$ e $r+dr$.

### Esercizio 07.4

L'obiettivo di questo esercizio è quello di equilibrare ed effettuare delle simulazioni **MC NVT** (ensemble canonico) per un modello di Lennard-Jones dell'Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) nelle seguenti condizioni:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

Per evitare di appesantire ulteriormente il notebook ho deciso di non mettere i plot relativi all'equilibrazione; nonostante questo faccio presente che per la fase solida e la fase liquida sono sufficienti $1000$ passi per l'equilibrazione, mentre la fase gassosa ne richiede almeno $3000$.

```python
U_S = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/potential_energy_S.dat", skiprows=1)
U_L = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/potential_energy_L.dat", skiprows=1)
U_G = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/potential_energy_G.dat", skiprows=1)
```

```python
dim=20
x = np.arange(dim)
x*=2000

plt.figure(figsize=(25, 4))
plt.errorbar(x,U_S[:, 2],yerr=U_S[:, -1])
plt.xlabel('#steps')
plt.ylabel('U/N')
plt.title('Solid phase - Potential energy per particle')
plt.grid(True)
plt.show()

plt.figure(figsize=(25, 4))
plt.errorbar(x,U_L[:, 2],yerr=U_L[:, -1], color='#00FF00')
plt.xlabel('#steps')
plt.ylabel('U/N')
plt.title('Liquid phase - Potential energy per particle')
plt.grid(True)
plt.show()

plt.figure(figsize=(25, 4))
plt.errorbar(x,U_G[:, 2],yerr=U_G[:, -1], color='red')
plt.xlabel('#steps')
plt.ylabel('U/N')
plt.title('Gas phase - Potential energy per particle')
plt.grid(True)
plt.show()
```

```python
P_S = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/pressure_S.dat", skiprows=1)
P_L = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/pressure_L.dat", skiprows=1)
P_G = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/pressure_G.dat", skiprows=1)
```

```python
plt.figure(figsize=(25, 4))
plt.errorbar(x,P_S[:, 2],yerr=P_S[:, -1])
plt.xlabel('#steps')
plt.ylabel('P')
plt.title('Solid phase - Pressure')
plt.grid(True)
plt.show()

plt.figure(figsize=(25, 4))
plt.errorbar(x,P_L[:, 2],yerr=P_L[:, -1], color='#00FF00')
plt.xlabel('#steps')
plt.ylabel('P')
plt.title('Liquid phase - Pressure')
plt.grid(True)
plt.show()

plt.figure(figsize=(25, 4))
plt.errorbar(x,P_G[:, 2],yerr=P_G[:, -1], color='red')
plt.xlabel('#steps')
plt.ylabel('P')
plt.title('Gas phase - Pressure')
plt.grid(True)
plt.show()
```

```python
g_S = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/gofr_S.dat", skiprows=1)
g_L = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/gofr_L.dat", skiprows=1)
g_G = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/gofr_G.dat", skiprows=1)

G_S = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/gofr_NVES.dat", skiprows=1)
G_L = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/gofr_NVEL.dat", skiprows=1)
G_G = np.loadtxt("Esercizio_07_4/nsl_simulator/NSL_SIMULATOR/OUTPUT/gofr_NVEG.dat", skiprows=1)
```

```python
plt.figure(figsize=(25, 4))
plt.errorbar(g_S[:, 0],g_S[:, 1],yerr=g_S[:, -1], label='MC NVT')
plt.errorbar(G_S[:, 0],G_S[:, 1],yerr=G_S[:, -1], label='MD NVE', linestyle='--')
plt.xlabel('r')
plt.ylabel('g(r)')
plt.title('Solid phase - Radial distribution function')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(25, 4))
plt.errorbar(g_L[:, 0],g_L[:, 1],yerr=g_L[:, -1], color='#00FF00', label='MC NVT')
plt.errorbar(G_L[:, 0],G_L[:, 1],yerr=G_L[:, -1], color='brown', label='MD NVE', linestyle='--')
plt.xlabel('r')
plt.ylabel('g(r)')
plt.title('Liquid phase - Radial distribution function')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(25, 4))
plt.errorbar(g_G[:, 0],g_G[:, 1],yerr=g_G[:, -1], color='red', label='MC NVT')
plt.errorbar(G_G[:, 0],G_G[:, 1],yerr=G_G[:, -1], color='yellow', label='MD NVE', linestyle='--')
plt.xlabel('r')
plt.ylabel('g(r)')
plt.title('Gas phase - Radial distribution function')
plt.grid(True)
plt.legend()
plt.show()
```

In questi tre plot ho confrontato i risultati ottenuti con simulazioni MC NVT per la funzione di distribuzione radiale $g(r)$  con quelli ottenuti con simulazioni **MD NVE** (ensemble microcanonico) nelle stesse condizioni termodinamiche, per ognuna delle tre fasi.

E' interessante notare che le due simulazioni agiscono in due ensemble termodinamici differenti. Tuttavia, grazie all'equilibrazione della temperatura nella simulazione MD NVE, i due sistemi vengono portati in condizioni termodinamiche che sono confrontabili, e che restituiscono quindi la stessa distribuzione radiale.
