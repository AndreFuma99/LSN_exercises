```python
def sum_1(t, Data):
    result = 0
    for i in range(int(tmax-t)):
        result += Data[i, 1]*Data[int(i+t), 1]
    return result

def sum_2(t, Data):
    result = 0
    for i in range(int(tmax-t)):
        result += Data[i, 1]
    return result

def sum_3(t, Data):
    result = 0
    for i in range(int(tmax-t)):
        result += Data[int(i+t), 1]
    return result

def sum_4(Data):
    result = 0
    for i in range(tmax):
        result += Data[i, 1]*Data[i, 1]
    return result

def sum_5(Data):
    result = 0
    for i in range(tmax):
        result += Data[i, 1]
    return result

def chi(t, Data):
    return (sum_1(t, Data)/(tmax-t)-(sum_2(t, Data)/(tmax-t))*(sum_3(t, Data)/(tmax-t)))/(sum_4(Data)/tmax-(sum_5(Data)/tmax)*(sum_5(Data)/tmax))

tmax = 500000
dim=100
CHI_S = np.zeros(dim)
CHI_L = np.zeros(dim)
CHI_G = np.zeros(dim)
T=np.linspace(0, tmax, num=dim)

for j in range(dim):
    CHI_S[j] = chi(T[j], Data_S)
    CHI_L[j] = chi(T[j], Data_L)
    CHI_G[j] = chi(T[j], Data_G)
```

```python
plt.figure(figsize=(25, 4))
plt.plot(T,CHI_S)
plt.xlabel('#steps')
plt.ylabel(r'$\chi$')
plt.title('Solid phase autocorrelation function')
plt.xlim(0, 10000)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(25, 4))
plt.plot(T,CHI_L)
plt.xlabel('#steps')
plt.ylabel(r'$\chi$')
plt.title('Liquid phase autocorrelation function')
plt.xlim(0, 10000)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(25, 4))
plt.plot(T,CHI_G)
plt.xlabel('#steps')
plt.ylabel(r'$\chi$')
plt.title('Gas phase autocorrelation function')
plt.xlim(0, 10000)
plt.grid(True)
plt.show()
```

```python
def error(AV,AV2,n):
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)
    
M=500000
dim=500
L=np.linspace(10, 7500, num=dim)
E=np.zeros(dim)

for m in range(dim):
    
    N=int(M/L[m])
    x = np.arange(N)
    ave = np.zeros(N)
    av2 = np.zeros(N)
    sum_prog = np.zeros(N)
    su2_prog = np.zeros(N)
    err_prog = np.zeros(N)
    
    for i in range(N):
        sum1 = 0
        for j in range(int(L[m])):
            k = j+i*int(L[m])
            sum1 += Data_S[k, 1]
        ave[i] = sum1/L[m] 
        av2[i] = (ave[i])**2

    for i in range(N):
        for j in range(i+1):
            sum_prog[i] += ave[j]
            su2_prog[i] += av2[j]
        sum_prog[i]/=(i+1)
        su2_prog[i]/=(i+1)
        err_prog[i] = error(sum_prog,su2_prog,i)
    
    E[m] = err_prog[N-1]
```

```python
plt.figure(figsize=(25, 4))
plt.plot(L,E)
plt.xlabel('#steps in each block')
plt.ylabel('Statistical uncertainty')
plt.title('Solid phase')
plt.axhline(y=np.mean(E[:333]), color='r')
plt.axhline(y=np.mean(E[333:]), color='g')
plt.grid(True)
plt.show()
```
