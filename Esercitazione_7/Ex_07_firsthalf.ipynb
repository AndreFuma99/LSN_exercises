### Esercizio 07.1

L'obiettivo di questo esercizio è quello di includere nel nostro codice le **tail corrections** per l'energia potenziale $U$ e la pressione $P$.

Le simulazioni infatti utilizzano frequentemente un potenziale con un cut-off sferico ad una distanza $r_c$. Diventa utile correggere i risultati delle simulazioni per compensare la parte persa di potenziale a lungo raggio.
Questi contributi sono frequentemente stimati assumendo la funzione di distribuzione radiale $g(r) \sim 1$ per $r>r_c$; per esempio il potenziale di Lennard-Jones dà (in unità ridotte):

$$
\frac{\langle V_{LJ} \rangle_{\text{tail}}}{N} = \frac{8 \pi \rho^{*}}{3} (\frac{1}{3r_{c}^{*9}} - \frac{1}{r_{c}^{*3}})
$$

La tail correction per la pressione è invece:

$$
\frac{\langle w_{LJ} \rangle_{\text{tail}}}{3N} = \frac{32 \pi \rho^{*}}{3} (\frac{1}{3r_{c}^{*9}} - \frac{1}{2r_{c}^{*3}})
$$

Il **tempo di autocorrelazione** $\chi(t)$ è una misura di quanto tempo impiega una variabile, in questo caso l'energia potenziale per particella $U/N$, per perdere memoria delle sue condizioni iniziali durante una simulazione Monte Carlo. In altre parole, indica quanto tempo (o quanti passi di simulazione) ci vuole affinché i valori di $U/N$ siano essenzialmente indipendenti l'uno dall'altro.
Nelle catene di Markov, il tempo nella simulazione Monte Carlo (MC) è discreto; pertanto, $\chi(t)$ può essere calcolato direttamente solo per un insieme finito di tempi MC. Se abbiamo un insieme di campioni di $U(t)$ misurati a intervalli di tempo regolari fino ad un tempo massimo $t_{max}$, allora la formula corretta per la funzione di autocorrelazione è:

$$
\chi(t) = \frac{1}{t_{\text{max}} - t} \left[ \sum_{t'=0}^{t_{\text{max}} - t} U(t') U(t' + t) - \sum_{t'=0}^{t_{\text{max}} - t} U(t') \cdot \sum_{t'=0}^{t_{\text{max}} - t} U(t' + t) \right]
\Bigg/
\left[
\frac{1}{t_{\text{max}}} \sum_{t'=0}^{t_{\text{max}}} U^2(t') - \left( \frac{1}{t_{\text{max}}} \sum_{t'=0}^{t_{\text{max}}} U(t') \right)^2
\right]
$$

```python
def sum_1(t, Data):
    result = 0
    for i in range(int(tmax-t)):
        result += Data[i, 1]*Data[int(i+t), 1]
    return result

def sum_2(t, Data):
    result = 0
    for i in range(int(tmax-t)):
        result += Data[i, 1]
    return result

def sum_3(t, Data):
    result = 0
    for i in range(int(tmax-t)):
        result += Data[int(i+t), 1]
    return result

def sum_4(Data):
    result = 0
    for i in range(tmax):
        result += Data[i, 1]*Data[i, 1]
    return result

def sum_5(Data):
    result = 0
    for i in range(tmax):
        result += Data[i, 1]
    return result

def chi(t, Data):
    return (sum_1(t, Data)/(tmax-t)-(sum_2(t, Data)/(tmax-t))*(sum_3(t, Data)/(tmax-t)))/(sum_4(Data)/tmax-(sum_5(Data)/tmax)*(sum_5(Data)/tmax))

tmax = 500000
dim=100
CHI_S = np.zeros(dim)
CHI_L = np.zeros(dim)
CHI_G = np.zeros(dim)
T=np.linspace(0, tmax, num=dim)

for j in range(dim):
    CHI_S[j] = chi(T[j], Data_S)
    CHI_L[j] = chi(T[j], Data_L)
    CHI_G[j] = chi(T[j], Data_G)
```

```python
plt.figure(figsize=(25, 4))
plt.plot(T,CHI_S)
plt.xlabel('#steps')
plt.ylabel(r'$\chi$')
plt.title('Solid phase autocorrelation function')
plt.xlim(0, 10000)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(25, 4))
plt.plot(T,CHI_L)
plt.xlabel('#steps')
plt.ylabel(r'$\chi$')
plt.title('Liquid phase autocorrelation function')
plt.xlim(0, 10000)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(25, 4))
plt.plot(T,CHI_G)
plt.xlabel('#steps')
plt.ylabel(r'$\chi$')
plt.title('Gas phase autocorrelation function')
plt.xlim(0, 10000)
plt.grid(True)
plt.show()
```

Osservando questi tre plot in cui viene rappresentato l'andamento della funzione di autocorrelazione $\chi$ in funzione del numero di steps per ognuna delle tre fasi, si nota che dopo circa $5000$ steps $\chi$ diventa sufficientemente piccolo: questo indica che dopo circa $5000$ steps i dati sono essenzialmente non correlati.

Ora voglio stimare le incertezze statistiche sul valore di aspettazione di $U/N$ per differenti dimensioni dei blocchi, da $L = M/N = 10$ a $L = 7500$, essendo $N$ il numero di blocchi.

```python
def error(AV,AV2,n):
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)
    
M=500000
dim=500
L=np.linspace(10, 7500, num=dim)
E=np.zeros(dim)

for m in range(dim):
    
    N=int(M/L[m])
    x = np.arange(N)
    ave = np.zeros(N)
    av2 = np.zeros(N)
    sum_prog = np.zeros(N)
    su2_prog = np.zeros(N)
    err_prog = np.zeros(N)
    
    for i in range(N):
        sum1 = 0
        for j in range(int(L[m])):
            k = j+i*int(L[m])
            sum1 += Data_S[k, 1]
        ave[i] = sum1/L[m] 
        av2[i] = (ave[i])**2

    for i in range(N):
        for j in range(i+1):
            sum_prog[i] += ave[j]
            su2_prog[i] += av2[j]
        sum_prog[i]/=(i+1)
        su2_prog[i]/=(i+1)
        err_prog[i] = error(sum_prog,su2_prog,i)
    
    E[m] = err_prog[N-1]
```

```python
plt.figure(figsize=(25, 4))
plt.plot(L,E)
plt.xlabel('#steps in each block')
plt.ylabel('Statistical uncertainty')
plt.title('Solid phase')
plt.grid(True)
plt.show()
```
