```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

```python
data = pd.read_csv("Es05.csv")
```

```python
print(data.head(8))
```

```python
dim=100000
x = np.arange(dim)
plt.figure(figsize=(25, 4))
plt.plot(x,data['Radius'])
plt.axhline(y=np.mean(data['Radius']), color="r")
plt.xlabel('#throws')
plt.ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
plt.grid(True)
plt.show()
```

```python
data1 = pd.read_csv("Es05_U0.csv")
```

```python
M=1000000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N)      # [0,1,2,...,N-1]

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,data1['CumulativeAverage'],yerr=data1['StatisticalUncertainty'])
plt.xlabel('#throws')
plt.ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
plt.grid(True)
plt.show()
```
