## <span style="color:blue">  Laboratorio di Simulazione Numerica - Esercitazione 05</span>

L'obiettivo di questa esercitazione è quello di utilizzare l'**algoritmo di Metropolis** per campionare il modulo quadro della funzione d'onda relativa allo stato fondamentale dell'atomo di idrogeno $|\Psi_{1,0,0}(x,y,z)|^2$ e relativa a uno dei tre stati eccitati $2p$ $|\Psi_{2,1,0}(x,y,z)|^2$ in coordinate cartesiane utilizzando prima una probabilità di transizione $T(\vec{x}|\vec{y})$ uniforme, poi una normale multivariata. Infine bisogna utilizzare queste posizioni campionate per stimare il valore atteso del raggio $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ e $\left\langle r \right\rangle_{\Psi_{2,1,0}}$.

L'obiettivo dell'algoritmo di Metropolis è generare una sequenza di campioni $X_1, X_2, ..., X_N$ che approssimano una distribuzione di probabilità target $\pi(x)$:

 - scelgo un punto iniziale $X_0$ nello spazio delle configurazioni
 - a partire da un punto corrente $X_t$ genero un punto candidato $Y$ secondo una distribuzione di probabilità di transizione $T(Y|X_t)$. Questa distribuzione può essere simmetrica o asimmetrica
 - calcolo la probabilità di accettazione $A$:
   $$
   A = \min\left(1, \frac{\pi(Y) T(X_t|Y)}{\pi(X_t) T(Y|X_t)}\right)
   $$
   se la probabilità di transizione è simmetrica, la probabilità di accettazione si semplifica a:
   $$
   A = \min\left(1, \frac{\pi(Y)}{\pi(X_t)}\right)
   $$
 - genero un numero casuale $s$ uniformemente distribuito nell'intervallo $[0,1]$. Se $s \leq A$, accetto la proposta e imposto $X_{t+1}=Y$; altrimenti, rifiuto la proposta e imposto $X_{t+1}=X_t$.
 - Ripeto questi passi un numero sufficiente di iterazioni fino ad ottenere una catena di campioni che sia rappresentativa della distribuzione target $\pi(x)$.

In coordinate cartesiane, il modulo quadro della funzione d'onda relativa allo stato fondamentale dell'atomo di idrogeno è:
$$
|\Psi_{1,0,0}(x,y,z)|^2 = \frac{a_{0}^{-3}}{\pi} e^{\frac{-2\sqrt{x^2+y^2+z^2}}{a_0}}
$$
se consideriamo invece uno dei tre stati eccitati $2p$, il modulo quadro della funzione d'onda è:
$$
|\Psi_{2,1,0}(x,y,z)|^2 = \frac{a_{0}^{-5}}{32\pi} z^2 e^{\frac{-\sqrt{x^2+y^2+z^2}}{a_0}}
$$

Il valore atteso del raggio risulta essere esattamente:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$

In seguito ho espresso le distanze in unità di raggio di Bohr $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529$ nm.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

```python
data_1 = pd.read_csv("T_Uniform/Ground_State/Es05_E0.csv")
```

```python
dim=100000
x = np.arange(dim)
plt.figure(figsize=(25, 4))
plt.plot(x,data_1['Radius'])
plt.axhline(y=np.mean(data_1['Radius']), color="r")
plt.xlabel('#throws')
plt.ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
plt.title('GS sampling equilibration - Uniform $T$')
plt.grid(True)
plt.show()
```

In questo plot ho rappresentato la fase di equilibrazione del campionamento di $|\Psi_{1,0,0}(x,y,z)|^2$ utilizzando una probabilità di transizione $T(\vec{x}|\vec{y})$ uniforme (in blu). Si può osservare che dopo circa $10^5$ tentativi il valore atteso del raggio $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ ha un valor medio circa pari al valore esatto $\frac{3}{2}$ (in rosso).

Si può osservare che dopo circa $10^5$ tentativi il valore atteso del raggio $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ ha un valor medio circa pari al valore esatto $5$ (in rosso).

```python
data_3 = pd.read_csv("T_Uniform/Ground_State/Es05_U0.csv")
data_4 = pd.read_csv("T_Uniform/2p_Excited_State/Es05_U1.csv")
```

```python
M=1000000             # Numero totale di lanci
N=100                 # Numero totale di blocchi
L=int(M/N)            # Number di lanci in ogni blocco
x = np.arange(N)      # [0,1,2,...,N-1]
x*=L # Numero di lanci = blocco * (Numero di lanci in ogni blocco)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].errorbar(x, data_3['CumulativeAverage'], yerr=data_3['StatisticalUncertainty'])
axs[0].set_xlabel('#throws')
axs[0].set_ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
axs[0].set_title('Ground state sampling - Uniform $T$')
axs[0].grid(True)

axs[1].errorbar(x, data_4['CumulativeAverage'], yerr=data_4['StatisticalUncertainty'])
axs[1].set_xlabel('#throws')
axs[1].set_ylabel(r'$\left\langle r \right\rangle_{\Psi_{2,1,0}}$')
axs[1].set_title('$2p$ excited state sampling - Uniform $T$')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

In questi plot ho rappresentato il campionamento di $|\Psi_{1,0,0}(x,y,z)|^2$ (a sinistra) e $|\Psi_{2,1,0}(x,y,z)|^2$ (a destra) utilizzando una probabilità di transizione $T(\vec{x}|\vec{y})$ uniforme. Si può osservare che all'aumentare del numero di "lanci" il valore atteso del raggio $\left\langle r \right\rangle_{\Psi}$ tende al valore esatto (rispettivamente $\frac{3}{2}$ e $5$). Le incertezze statistiche sono state stimate utilizzando data blocking.

### Osservazioni:

Ho scelto come step della probabilità di transizione $T(\vec{x}|\vec{y})$, in modo da ottenere il $50$% di accettazione in tutti i casi
 - $T$ uniforme, stato fondamentale: step = $1.227$
 - $T$ uniforme, stato eccitato $2p$: step = $2.98$
 - $T$ gaussiana, stato fondamentale: step = $0.755$
 - $T$ gaussiana, stato eccitato $2p$: step = $1.88$

Il $50$% è un valore ottimale per la probabilità di accettazione perchè assicura che una frazione significativa delle proposte di transizione venga accettata, permettendo al Metropolis di esplorare delle nuove regioni dello spazio delle configurazioni. Inoltre, il $50$% implica che le proposte di transizione sono sufficientemente grandi da evitare che l'algoritmo faccia solo piccoli passi, mantenendo l'efficienza del campionamento.

E' possibile osservare che, partendo con un punto iniziale molto lontano dall'origine e osservando il processo di equilibrazione, il Metropolis inizialmente ha difficoltà a trovare configurazioni accettabili, ma gradualmente si avvicina alle regioni di spazio dove la distribuzione di probabilità è significativa.

Io ho effettuato $10^6$ Monte Carlo Metropolis steps e li ho suddivisi in 100 blocchi, ognuno contenente quindi $10^4$ campioni. Per verificare che questa configurazione produca un campionamento efficiente con l'algoritmo di Metropolis, si potrebbe calcolare il tempo di autocorrelazione $\tau$: un blocco dovrebbe contenere un numero di campioni significativamente maggiore di $\tau$ per assicurarsi che la maggior parte dei campioni nel blocco sia indipendente.

Infine osservo che, implementando l'algoritmo di Metropolis con una probabilità di transizione normale multivariata i risultati per $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ e $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ sono equivalenti a quelli ottenuti utilizzando una probabilità di transizione uniforme: è stato necessario garantire una buona fase di equilibrazione, una scelta appropriata dello step di $T(\vec{x}|\vec{y})$ e un numero sufficiente di campioni e blocchi.
