```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

```python
data_1 = pd.read_csv("T_Uniform/Ground_State/Es05_E0.csv")
```

```python
dim=100000
x = np.arange(dim)
plt.figure(figsize=(25, 4))
plt.plot(x,data_1['Radius'])
plt.axhline(y=np.mean(data_1['Radius']), color="r")
plt.xlabel('#throws')
plt.ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
plt.title('GS sampling equilibration')
plt.grid(True)
plt.show()
```

```python
data_2 = pd.read_csv("T_Uniform/Ground_State/Es05_U0.csv")
```

```python
M=1000000              # Total number of throws
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N)      # [0,1,2,...,N-1]

x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,data_2['CumulativeAverage'],yerr=data_2['StatisticalUncertainty'])
plt.xlabel('#throws')
plt.ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
plt.title('Ground state sampling')
plt.grid(True)
plt.show()
```

```python
data_3 = pd.read_csv("T_Uniform/Ground_State/Es05_U0_xyz.csv")
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')
ax.scatter(data_3['X'], data_3['Y'], data_3['Z'], c=data_3['Z'], marker='.', s=1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)

plt.show()
```
