```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points=100
T = np.linspace(0.2,2.5,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

x = [2.0, 1.75, 1.5, 1.25, 1.0, 0.75, 0.5]
U_metro = [-0.47044, -0.516348, -0.582653, -0.664277, -0.762102, -0.870884, -0.974108]
U_gibbs = [-0.47007, -0.516125, -0.582512, -0.665071, -0.761672, -0.870497, -0.971538]
eU_metro = [0.0079036, 0.000265682, 0.000284111, 0.000212158, 0.000350045, 0.00049811, 0.000853757]
eU_gibbs = [0.00788161, 0.000272669, 0.000348265, 0.000358296, 0.000369961, 0.00076436, 0.00100843]


plt.plot(T, e)
plt.scatter(x, U_metro, color='green')
plt.scatter(x, U_gibbs, color='red')
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.show()
```

```python
U_M = [None] * len(x)
U_G = [None] * len(x)
for i in range (len(x)):
    th = np.tanh(J/x[i])
    thN= th**Ns
    ch = 1/th
    e = -J*( th + ch*thN )/( 1 + thN )
    U_M[i] = e-U_metro[i]
    U_G[i] = e-U_gibbs[i]
    
plt.errorbar(x, U_M, yerr=eU_metro, label='Metropolis')
plt.errorbar(x, U_G, yerr=eU_gibbs, label='Gibbs')
plt.xlabel('T')
plt.ylabel(r'$(U_{\mathrm{th}} - U_{\mathrm{sim}})/N$')
plt.title('Theory vs Simulation: Internal Energy')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
points=100
T = np.linspace(0.2,2.5,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

C_metro = [0.195988, 0.234703, 0.287689, 0.350433, 0.410782, 0.420612, 0.19786]
C_gibbs = [0.196068, 0.23475, 0.287764, 0.349765, 0.411532, 0.421912, 0.217448]
eC_metro = [0.00327501, 8.7484e-05, 0.000143372, 0.000185758, 0.00052654, 0.00156661, 0.00652142]
eC_gibbs = [0.00328493, 9.29483e-05, 0.000184545, 0.000298723, 0.000556555, 0.00238918, 0.00769313]

plt.plot(T, heat)
plt.scatter(x, C_metro, color='green')
plt.scatter(x, C_gibbs, color='red')
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.show()
```

```python
C_M = [None] * len(x)
C_G = [None] * len(x)
for i in range (len(x)):
    beta = 1/x[i]
    th = np.tanh(J/x[i])
    thN= th**Ns
    ch = 1/th
    heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
    C_M[i] = heat-C_metro[i]
    C_G[i] = heat-C_gibbs[i]
    
plt.errorbar(x, C_M, yerr=eC_metro, label='Metropolis')
plt.errorbar(x, C_G, yerr=eC_gibbs, label='Gibbs')
plt.xlabel('T')
plt.ylabel(r'$C_{\mathrm{th}} - C_{\mathrm{sim}}$')
plt.title('Theory vs Simulation: Heat Capacity')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
