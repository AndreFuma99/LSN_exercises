```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import scipy.stats
```

```python
data = np.loadtxt("/home/andrea/Documenti/CorsoLSN/Esercitazione_6/nsl_simulator/NSL_SIMULATOR/OUTPUT/equilibration.dat", skiprows=1)
```

```python
N=7000               # Number of blocks
L=1            # Number of steps in each block
x = np.arange(N)      # [0,1,2,...,N-1]

x*=L # Number of steps = block * (Number of steps in each block)
plt.plot(x, data[:, 2])
plt.xlabel('#steps')
plt.ylabel('U*')
plt.title('Equilibration')
plt.grid(True)
plt.show()
```

```python
U = []
C = []
X = []
M = []

eU = []
eC = []
eX = []
eM = []
```

```python
dataU = np.loadtxt("/home/andrea/Documenti/CorsoLSN/Esercitazione_6/nsl_simulator/NSL_SIMULATOR/OUTPUT/total_energy.dat", skiprows=1)
dataC = np.loadtxt("/home/andrea/Documenti/CorsoLSN/Esercitazione_6/nsl_simulator/NSL_SIMULATOR/OUTPUT/specific_heat.dat", skiprows=1)
dataX = np.loadtxt("/home/andrea/Documenti/CorsoLSN/Esercitazione_6/nsl_simulator/NSL_SIMULATOR/OUTPUT/susceptibility.dat", skiprows=1)
```

```python
U.append(dataU[-1, 2])
C.append(dataC[-1, 2])
X.append(dataX[-1, 2])

eU.append(dataU[-1, -1])
eC.append(dataC[-1, -1])
eX.append(dataX[-1, -1])
```

```python
print(U)
print(C)
print(X)
```

```python
print(eU)
print(eC)
print(eX)
```

```python
x = [2.0, 1.75, 1.5, 1.25, 1.0, 0.75, 0.5]

plt.errorbar(x, U, yerr=eU)
plt.xlabel('T')
plt.ylabel('U/N')
plt.title('Ising 1D, internal energy')
plt.grid(True)
plt.show()
```

```python
plt.errorbar(x, C, yerr=eC)
plt.xlabel('T')
plt.ylabel('C')
plt.title('Ising 1D, heat capacity')
plt.grid(True)
plt.show()
```

```python
plt.errorbar(x, X, yerr=eX)
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.title('Ising 1D, susceptibility')
plt.grid(True)
plt.show()
```
